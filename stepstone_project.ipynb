```python
import pandas as pd
import matplotlib.pyplot as plt

# Re-creating the DataFrame with the skill data provided
skills_data = {
    'Skill': ['SQL', 'Python', 'MS Excel', 'Microsoft BI', 'Cloud technologies',
              'R', 'Tableau', 'Data Visualization', 'Data Analysis'],
    'Occurrences': [227, 148, 148, 106, 80, 76, 56, 56, 38]  # Occurrences are listed for descending visualization
}
skills_df = pd.DataFrame(skills_data)

# Sorting the DataFrame in descending order for better visualization
sorted_skills_df = skills_df.sort_values(by='Occurrences', ascending=False)

# Displaying the sorted DataFrame
print(sorted_skills_df)  # Use print() to ensure the DataFrame is outputted

# Plotting
plt.figure(figsize=(12, 8))
plt.barh(sorted_skills_df['Skill'], sorted_skills_df['Occurrences'], color='lightgreen')
plt.xlabel('Occurrences')
plt.ylabel('Skills')
plt.title('Demand for Data Analyst Skills in Job Listings (Descending Order)')
plt.gca().invert_yaxis()  # To display the highest value on top
plt.show()


   
```
